```python
data_sorted = data.sort_values(by="Температура тела")
tails = data_sorted["Длина головяных отростков"]
temperature = data_sorted["Температура тела"]
```

```python
fig, axes = plt.subplots(1, 1, figsize=(4, 4))
axes.scatter(tails, temperature, label="Температура")
plt.legend(loc=0)
plt.show()
```

```python
tails_dichotomy = data_sorted[data_sorted["Диагноз"]=="Дихотомия"]["Длина головяных отростков"]
temperature_dichotomy = data_sorted[data_sorted["Диагноз"]=="Дихотомия"]["Температура тела"]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(tails, temperature, label="Температура")
axes[1].scatter(tails_dichotomy, temperature_dichotomy, label="Температура")
plt.legend(loc=0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
axes[0].scatter(tails, temperature, label="Температура")
axes[1].scatter(tails_dichotomy, temperature_dichotomy, label="Температура")
plt.legend(loc=0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
axes[0].scatter(tails, temperature, label="Температура")
axes[1].scatter(tails_dichotomy, temperature_dichotomy, label="Температура")
axes[1].plot(tails_dichotomy, 2*tails_dichotomy, label="y=2x", color="red")
plt.legend(loc=0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
axes[0].scatter(tails, temperature, label="Температура")
axes[0].set_title("Все фарсиане")
axes[1].scatter(tails_dichotomy, temperature_dichotomy, label="Температура")
axes[1].plot(tails_dichotomy, temperature_dichotomy, label="y=2x", color="red")
axes[1].set_title("Только больные трихотомией")
plt.legend(loc=0)
plt.show()
```
