# I. Numpy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
array_1D = np.zeros(10)
array_1D[4] = 1
array_2D = array_1D.reshape(1, 10)
print(array_2D)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
array_1D = np.arange(10, 50)
print(array_1D)

array_1D_rev = array_1D[::-1]
print(array_1D_rev)

array_1D_rev_even = array_1D_rev[array_1D_rev % 2 == 0]
print(array_1D_rev_even)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
array_3D = np.random.randint(0, 60, size = (4, 3, 2))
#array_3D = 20 * np.random.rand(4, 3, 2)
min_value = np.min(array_3D)
max_value = np.max(array_3D)

print(array_3D)
print(min_value)
print(max_value)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
array_2D_1 = np.random.randint(0, 20, size = (6, 4))
array_2D_2 = np.random.randint(20, 40, size = (4, 3))
array = array_2D_1 @ array_2D_2
print(array)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array_2D = np.random.randint(0, 100, size = (7, 7))
print(array_2D)

mean = np.mean(array_2D)
print(f"\nMean: {mean}")

std = np.std(array_2D)
print(f"std: {std}\n")

array_norm = (array_2D - mean) / std
print(array_norm)

mean = np.mean(array_norm)
print(f"\nMean: {mean} -> {round(mean)}")

std = np.std(array_norm)
print(f"std: {std}\n")
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
data = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
print(data.head(5))
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(data.shape)
print(data.columns)
```

### Проверьте есть ли пропуски в данных

```python
print( data.isnull().sum() )
print( data.isnull().sum().sum() )
print( data.isnull().values.any() )
print( data.info() )
```

### Посмотрите на распределение числовых признаков

```python
print( data.describe() )

data.hist(bins=15, figsize=(3, 3))
plt.tight_layout()
plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
print( data["total_bill"].max() )

print( data["total_bill"].nlargest(1) )

print( data[data["total_bill"] == data["total_bill"].max()] )

print( data["total_bill"].describe() )
```

### Найдите количество курящих людей

```python
print( data[data["smoker"] == "Yes"].shape[0] )

print( len(data[data["smoker"] == "Yes"]) )

print( (data["smoker"] == "Yes").sum() )

print( data["smoker"].value_counts() )
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print( data["day"].value_counts() )

print( data.groupby("day")["total_bill"].mean() )

print( data.pivot_table(values="total_bill", index="day", aggfunc = "mean") )

sns.barplot(x="day", y="total_bill", data=data, estimator="mean")
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
#print( data["total_bill"].median() )
filt_data_bill = data[data["total_bill"] > data["total_bill"].median()]
#print(filt_data_bill["total_bill"])

print( filt_data_bill.groupby("sex")["tip"].mean() ) 

print( filt_data_bill.pivot_table(values="tip", index="sex", aggfunc="mean") ) 

sns.barplot(data=filt_data_bill, x="sex", y="tip", estimator="mean")
plt.show()
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
data["smoker_bin_1"] = data["smoker"].map({"No":0, "Yes":1})
print( data[["smoker", "smoker_bin_1"]].iloc[65:70] )

data["smoker_bin_2"] = data["smoker"].apply(lambda x: 1 if x == "Yes" else 0)
print( data[["smoker", "smoker_bin_1"]].iloc[65:70] )
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
data["total_bill"].hist(bins = 25, figsize = (3, 3))
plt.xlabel("Всего чеков")
plt.ylabel("Размер чека")
plt.title("Распределение")
plt.show()

plt.figure(figsize=(3, 3))
sns.histplot(data=data, x="total_bill", bins=25, hue="sex", kde=True)
plt.xlabel("Всего чеков")
plt.ylabel("Размер чека")
plt.title("Распределение")
plt.show()

plt.figure(figsize=(3, 3))
sns.histplot(data=data, x="total_bill", bins=25, hue="smoker", kde=True)
plt.xlabel("Всего чеков")
plt.ylabel("Размер чека")
plt.title("Распределение")
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(3, 3))
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex")
plt.xlabel("Сумма чека")
plt.ylabel("Сумма чаевых")
plt.show()

plt.figure(figsize=(3, 3))
plt.scatter(data["total_bill"], data["tip"])
plt.xlabel("Сумма чека")
plt.ylabel("Сумма чаевых")
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(data, hue="smoker", palette="muted", height=1.5)
plt.show()

sns.pairplot(data, kind="kde",  hue="sex", palette="muted", height=1.5)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'  

```python
plt.figure(figsize=(3, 3))
sns.barplot(data=data, x='day', y='total_bill', estimator='median')
plt.show()

plt.figure(figsize=(3, 3))
sns.boxplot(data=data, x="day", y="total_bill")
plt.show()

plt.figure(figsize=(3, 3))
sns.violinplot(data=data, x="day", y="total_bill")
plt.show()

plt.figure(figsize=(3, 3))
sns.stripplot(data=data, x='day', y='total_bill')
plt.show()

plt.figure(figsize=(3, 3))
sns.swarmplot(data=data, x='day', y='total_bill')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(3, 3))
sns.histplot(data=data, x="day", hue="time")
plt.show()

graph = sns.FacetGrid(data, col='time')
graph.map_dataframe(sns.histplot, 'tip', bins=10, kde=True)
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
plt.figure(figsize=(3, 3))
graph = sns.FacetGrid(data=data, col="sex", hue="smoker")
graph.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
graph.add_legend()
plt.show()
```

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', col='sex', kind='scatter', height=3)
plt.show()
```
